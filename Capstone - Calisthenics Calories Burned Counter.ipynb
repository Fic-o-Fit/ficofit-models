```python
import numpy as np
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([                     
        tf.keras.layers.Dense(units=1, input_shape=[1])  
])
# input = tf.keras.Input(shape=(1))
# layer = tf.divide(input, [100.0])
# output = tf.keras.layers.Dense(units=1)(layer)

# model = tf.keras.Model(input, output, name="calorieCounter")

model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([0.0, 0.59, 0.70, 0.82, 0.93], dtype=np.float32) #the weight is divided by 100
# xs = np.array([0.0, 59, 70, 82, 93], dtype=np.float32)
ys = np.array([0.0, 0.39, 0.47, 0.55, 0.62], dtype=np.float32)

# xs = np.array([0.0, 58.96, 70.30, 81.64, 92.98], dtype=np.float32)
# ys = np.array([0.0, 39.33, 46.91, 54.50, 62.08], dtype=np.float32)
model.fit(xs, ys, epochs=500)
```

```python
model.summary()
```

```python
print(model.predict([0.5]))
```

```python
saved_model_path = "./calorie-counter.h5"

model.save(saved_model_path)
```
